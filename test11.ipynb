### Utility functions

```python
def ARtoDMG(AR,DEF=DEF["AVG"],ABS=ABS):
    """
    Converts an AR array to a real damage array
    Parameters:
        AR: list or pandas Series/DataFrame of length 8
            AR for each damage type.
        DEF: list or pandas Series/DataFrame of length 8
            Defenses for each damage type. Default values found in earlier section.
        ABS: list or pandas Series/DataFrame of length 8
            Absorbtions for each damage type. Default values found in earlier section.
    Output:
        pandas series of length 8
    """
    res=[]
    for x,y,z in zip(AR,DEF,ABS):
        if y>x*8:
            res.append((1-z)*0.1*x) 
        elif y>x:
            res.append((1-z)*(19.2/49*(x/y-0.125)**2+0.1)*x)
        elif y>x*0.4:
            res.append((1-z)*(-0.4/3*(x/y-2.5)**2+0.7)*x)
        elif y>x/8:
            res.append((1-z)*(-0.8/121*(x/y-8)**2+0.9)*x)
        else:
            res.append((1-z)*0.9*x)
    return pd.Series(res)
```

```python
def DMGweapon(weapon,counterHits=True):
    """
    Computes real damage and if relevant real pierce and real spear tali pierce damage
    Parameters:
        weapon: Multindexed pandas DataFrame with 8 rows
            Row index level 0 is weapon name. Row index level 1 is damage type.
            Column index level 0 is build name. Column index level 1 is infusion/buff. See example in the Weapons section.
        counterHits: boolean
            Calculate counter hits damage and spear talisman counter hit damage if the weapon deals pierce damage. Adds 2 rows to the output.
    Output:
        Multindexed pandas DataFrame of either 1 or 3 rows.
            Row index level 0 is weapon name. Row index level 1 is damage category (real, prc, prc spr). Same column index as weapon.
    """
    real=weapon.apply(ARtoDMG).set_index(weapon.index).groupby(level=0).sum().applymap(int)
    if not counterHits:
        return real
    if (weapon.iloc[3,:]==0).all():
        return pd.concat([real],keys=["Real"]).swaplevel()
    realprc=weapon.apply(ARtoDMG).set_index(weapon.index)
    realprc.loc[pd.IndexSlice[:, 'Pierce'], :]*=1.15
    realprc=realprc.groupby(level=0).sum().applymap(int)
    realprcspr=weapon.apply(ARtoDMG).set_index(weapon.index)
    realprcspr.loc[pd.IndexSlice[:, 'Pierce'], :]*=1.15*1.15
    realprcspr=realprcspr.groupby(level=0).sum().applymap(int)
    return pd.concat([real,realprc,realprcspr],keys=["Real","Prc","Prc Spr"]).swaplevel()
```

```python
def statsFormatter(x):
    """To show build stats in column header"""
    for b in BUILDS:
        x=x.replace(b,b+" - "+" ".join(BUILDS[b][3:].astype(str).tolist()))
    return x

def DMGtable(weapons,weaponBuffs=True,counterHits=True,percentage=True,showStats=True,multicolor=True):
    """
    Displays a fancy damage table.
    Parameters:
        weapons: list of Multindexed pandas DataFrame with 8 rows
            Row index level 0 is weapon name. Row index level 1 is damage type.
            Column index level 0 is build name. Column index level 1 is infusion/buff. See example in the Weapons section.
        weaponBuffs: boolean
            Display weapon buffs like greases or Flaming Strike/Sacred Blade.
        counterHits: boolean
            Display counter damage and spear tali counter damage.
        percentage: boolean
            Display the difference ratio between cell damage and max row damage. Buffs not taken into account for max row damage.
            For example -x% means x% less damage than the best non buffed infusion. 0% means this is the best unbuffed infusion. x% means this is the best buffed infusion.
        showStats: boolean
            Display build stat spread in column header.
        multicolor:
            Display one color per build.
    """
    baseInfusions=["Hvy","Fire","Keen","Ltng","Mgc","Cold","Scrd","Blood","Occlt"]
    weapons=pd.concat([DMGweapon(w,counterHits=counterHits) for w in weapons]).astype("Int64")
    if not weaponBuffs:
        weapons=weapons.loc[:,pd.IndexSlice[:, baseInfusions]]
    weaponsRatio=weapons.apply(lambda x:x/x.loc[pd.IndexSlice[:, baseInfusions]].max()*100-100,axis=1).astype(float)
    tmp=weapons.copy()
    for c in weapons.columns:
        if percentage:
            tmp[c]=weapons[c].map(lambda x:str(x).replace("<NA>","-"))+" ("+weaponsRatio[c].round(2).map(lambda x:("👑" if x==0 else (("+" if x>=0 else "-")+str(abs(x)).replace("nan",""))+"%"))+")"
        else:
            tmp[c]=weapons[c].map(lambda x:str(x).replace("<NA>","-"))
    res=tmp.style
    for i in tmp.index:
        vmin=-20#weaponsRatio.loc[[i]].min().min()-0.01
        if not multicolor:
            res.background_gradient(cmap="Greens",axis=None,gmap=weaponsRatio,vmin=vmin,vmax=0)
        else:
            res.background_gradient(
                cmap=LinearSegmentedColormap.from_list("",["white","red"]),axis=None,gmap=weaponsRatio,subset=(i,tmp[["STR"]].columns),vmin=vmin,vmax=2).background_gradient(
                    cmap=LinearSegmentedColormap.from_list("",["white","gold"]),axis=None,gmap=weaponsRatio,subset=(i,tmp[["DEX"]].columns),vmin=vmin,vmax=2).background_gradient(
                        cmap=LinearSegmentedColormap.from_list("",["white","blue"]),axis=None,gmap=weaponsRatio,subset=(i,tmp[["INT"]].columns),vmin=vmin,vmax=2).background_gradient(
                            cmap=LinearSegmentedColormap.from_list("",["white","orange"]),axis=None,gmap=weaponsRatio,subset=(i,tmp[["FTH"]].columns),vmin=vmin,vmax=2).background_gradient(
                                cmap=LinearSegmentedColormap.from_list("",["white","violet"]),axis=None,gmap=weaponsRatio,subset=(i,tmp[["ARC"]].columns),vmin=vmin,vmax=2)
    for i in [j for i,j in zip(tmp.columns[:-1],tmp.columns[1:]) if i[0]!=j[0]]:
        res.set_table_styles({i: [{'selector': '', 'props': 'border-left: 1px solid black !important;'}]}, overwrite=False)
    if counterHits:
        for i in [j for i,j in zip(tmp.index[:-1],tmp.index[1:]) if i[0]!=j[0]]:
            res.set_table_styles({i: [{'selector': '', 'props': 'border-top: 1px solid black !important;'}]}, overwrite=False, axis=1)
    if showStats:
        res.format_index(statsFormatter,axis=1)
    display(res)
```
